```python
# Importa pandas como pd
import pandas as pd
import matplotlib.pyplot as plt
import datetime
import numpy as np
# Importa '../data/car-sales.csv' en un DataFame llamado car_sales y visualízalo
car_sales = pd.read_csv("car-sales.csv")
# Intenta trazar la columna 'Precio' usando la función plot()
#plt.plot(car_sales['Price'])
plt.show()
print(car_sales)
```

¿Por qué no funciona?

Pista: No son datos numéricos.

En el proceso de convertirlos en datos numéricos, creemos otra columna que sume la cantidad total de ventas y otra que muestre en qué fecha se vendió el automóvil.

Pista: Para sumar una columna de forma acumulativa, busca la función cumsum(). Y para crear una columna de fechas, busca la función date_range().

```python
#car_sales['Total'] = car_sales.cumsum()['Price']
#car_sales['Fechas'] = car_sales.date_range(start='2020-01-01', end='2020-01-10')

# Quitar los símbolos, los dos últimos números de la columna 'Precio' y convertirlos a números

car_sales['Price']=car_sales['Price'].str.replace('$', '')
car_sales['Price']=car_sales['Price'].str.replace('.00', '')
car_sales['Price']=car_sales['Price'].str.replace(',', '')
#car_sales['Price']=car_sales.to_numeric(car_sales['Price'], errors='coerce')
car_sales['Price']=car_sales['Price'].astype(int)
print(car_sales['Price'].dtype)
car_sales
```

```python
# Agregar una columna llamada 'Ventas totales' a car_sales que suma de forma acumulativa la columna 'Precio'
car_sales['Total'] = car_sales.cumsum()['Price']
car_sales
x=pd.date_range("06-03-2025", periods=10, freq="ME")



# Agregar una columna llamada 'Fecha de venta' que enumera una serie de fechas sucesivas a partir de hoy (su hoy)
#car_sales['Fecha de venta'] = car_sales.date_range("2018-01-01", periods=3, freq="h")
car_sales['Fecha de venta']= x
# Ver el DataFrame de car_sales
car_sales
```

Ahora tenemos una columna numérica (Ventas totales) y una columna de fechas (Fecha de venta), visualicémoslas.

```python
print(car_sales.loc[:,['Total','Fecha de venta']])
# Usar la función plot() para trazar la columna 'Fecha de venta' en comparación con la columna 'Ventas totales'
x=car_sales['Total']
y=car_sales['Fecha de venta']
fig, ax = plt.subplots()
ax.plot(x,y)
plt.show()
```

```python
# Convertir la columna 'Precio' a números enteros
car_sales['Price']=car_sales['Price'].astype(int)
print(car_sales['Price'].dtype)
# Crear un diagrama de dispersión de las columnas 'Odómetro (KM)' y 'Precio' utilizando la función plot()
x=car_sales['Odometer (KM)']
y=car_sales['Price']
fig, ax = plt.subplots()
ax.scatter(x,y)
plt.show()
```
